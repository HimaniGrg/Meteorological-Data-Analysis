#Weather Suitability Analysis

```python
# Define weights for each weather parameter (adjust as needed)
weight_temperature = 0.3
weight_precipitation = 0.4    # prioritize precipitation

# Define suitability thresholds (adjust as needed)
suitable_threshold = 0.6
moderate_threshold = 0.4
```

```python
# Create an empty DataFrame to store the results
results_df = pd.DataFrame(columns=['CITY','YEAR','MONTH','DATE','Temperature_Score', 'Precipitation_Score', 'Suitability_Score', 'Weather_Class'])
```

```python
# function to calculate weather suitability score
def calculate_suitability_score(city_data):
    # # Create an empty DataFrame to store the results
    # results_df = pd.DataFrame(columns=['CITY','YEAR','MONTH','DATE','Temperature_Score', 'Precipitation_Score', 'Suitability_Score', 'Weather_Class'])

    # Normalize parameters to a scale between 0 and 1 (min-max normalization)
  city_data['Temperature_Score'] = (city_data['Temp_mean'] - city_data['Temp_mean'].min()) / (city_data['Temp_mean'].max() - city_data['Temp_mean'].min())
  city_data['Precipitation_Score'] = 1 - (city_data['Precipitation'] / city_data['Precipitation'].max())

    # Calculate the suitability score for each day in the city
  city_data['Suitability_Score'] = (
        weight_temperature * city_data['Temperature_Score'] +
        weight_precipitation * city_data['Precipitation_Score']
    )

  # Classify days based on suitability score
  city_data['Weather_Class'] = np.where(city_data['Suitability_Score'] >= suitable_threshold, 'Suitable',
                                          np.where(city_data['Suitability_Score'] >= moderate_threshold, 'Moderate', 'Not Suitable'))

  return city_data
```

```python
# Iterate over each city's data
for city, city_data in df.groupby('CITY'):
  city_data_df = calculate_suitability_score(city_data)

   # Append the results for the current city to the results DataFrame
  results_df = pd.concat([results_df, city_data_df[['CITY', 'YEAR','MONTH','DATE', 'Temperature_Score', 'Precipitation_Score', 'Suitability_Score', 'Weather_Class']]])
```

#Train a model

Task: predict weather (Suitable/ Moderate/ Not suitable) for the next day based on single-day weather observations.

Technique: Random forest classifier

Tool: Python + Scikit-learn

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Initialize the label encoder
label_encoder = LabelEncoder()

# Apply label encoding to the 'City' and 'Country' column
df['City_encoded'] = label_encoder.fit_transform(df['CITY'])
df['Country_encoded'] = label_encoder.fit_transform(df['COUNTRY'])
# Display the encoded DataFrame
df
```

```python
country_label = dict(zip(df['COUNTRY'],df['Country_encoded']))
print(country_label)
```

```python
city_label = dict(zip(df['CITY'],df['City_encoded']))
print(city_label)
```

```python
# Assuming you have features in X and target in y
X = df.drop(['Weather_Class','CITY','COUNTRY'], axis=1)
y = df['Weather_Class']
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", y_train.shape)
print("Y_test shape:", y_test.shape)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust the number of trees (n_estimators)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
type(y_test)
```

```python
type(y_pred)
```

```python
# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test,y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Print the accuracy and classification report
print(f'Accuracy: {accuracy}')
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print('Classification Report:\n', report)
```

Above evaluation shows that most of the days are classified as moderate while few of them are classified as Not Suitable. This means most of the time through the decade was suitable for picnic.
